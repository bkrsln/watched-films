```python
import pandas as pd
import numpy as np
from bokeh.plotting import figure, output_file, output_notebook, show
from bokeh.models import ColumnDataSource

df = pd.read_csv('films.csv', encoding = "ISO-8859-1", parse_dates=['Date Rated'])
```

```python
df.head()
```

## Filmin çıkış yılına göre benim ve IMDb'deki puanları

```python
rating = ColumnDataSource(df)

p = figure(x_axis_label = 'Release Year', y_axis_label = 'Rating Score')

p.circle(x='Year', y='IMDb Rating', source=rating, legend_label='IMDb Rating', size=10, color="green", alpha=0.7)
p.circle(x='Year', y='Your Rating', source=rating, legend_label='My Rating', size=8, color="red", alpha=0.4)
p.legend.location = "bottom_left"
p.title.text = "Films (Release Year / Rating Score)"

output_notebook()
show(p)
```

## IMDb sitesini veya uygulamasını sık kullandığım yıllar

```python
rating = ColumnDataSource(df)

p = figure(x_axis_type='datetime',
           x_axis_label = 'Date Rated',
           y_axis_label = 'Rating Score')

p.circle('Date Rated', 'Your Rating', source=rating, legend_label='My Rating', size=10, color="red", alpha=0.5)
p.legend.location = "bottom_left"
p.title.text = "Films (Date Rated / Rating Score)"

output_notebook()
show(p)
```

```python
df.Year = df.Year.astype(str)

year_group = df.groupby(('Year'))

year = ColumnDataSource(year_group)

year_group.describe()
```

## Hangi yıllara ait filmleri daha çok izledim?

```python
p = figure(x_range=year_group, plot_height=500)

p.vbar(x='Year', top='Position_count', width=0.7, line_color="black", color="orange", alpha=0.7, source=year)

p.xaxis.axis_label = "Release Year"
p.yaxis.axis_label = "Film Count"
p.y_range.start = 0
p.xaxis.major_label_orientation = 3.14/2
p.title.text = "Films (Release Year / Film Count)"

output_notebook()
show(p)
```

## En yüksek ve en düşük puan verdiğim yıllar hangileri?

```python
p = figure(x_range=year_group, plot_height=500)

p.line(x='Year', y='Your Rating_mean', width=1.5, line_color="navy", alpha=0.8, source=year, 
         legend_label="My Rating (Mean)")
p.circle(x='Year', y='Your Rating_max', size=5, color="green", alpha=0.8, source=year, 
         legend_label="My Rating (Max)")
p.circle(x='Year', y='Your Rating_min', size=5, color="red", alpha=0.8, source=year, 
         legend_label="My Rating (Min)")

p.xaxis.axis_label = "Release Year"
p.yaxis.axis_label = "My Rating"
p.y_range.start = 3
p.xaxis.major_label_orientation = 3.14/2
p.title.text = "Films (Release Year / My Rating)"
p.legend.location = "bottom_left"

output_notebook()
show(p)
```

## IMDb'deki puanların ortalaması ne?

```python
p = figure(x_range=year_group, plot_height=500)

p.line(x='Year', y='IMDb Rating_mean', width=1.5, line_color="navy", alpha=0.8, source=year, 
         legend_label="IMDb Rating (Mean)")
p.circle(x='Year', y='IMDb Rating_max', size=5, color="green", alpha=0.8, source=year, 
         legend_label="IMDb Rating (Max)")
p.circle(x='Year', y='IMDb Rating_min', size=5, color="red", alpha=0.8, source=year, 
         legend_label="IMDb Rating (Min)")

p.xaxis.axis_label = "Release Year"
p.yaxis.axis_label = "IMDb Rating"
p.y_range.start = 3
p.xaxis.major_label_orientation = 3.14/2
p.title.text = "Films (Release Year / IMDb Rating)"
p.legend.location = "bottom_left"

output_notebook()
show(p)
```

## Benim ve IMDb puanları arasındaki ilişki nasıl?

```python
p = figure(x_range=year_group, plot_height=500)

p.line(x='Year', y='IMDb Rating_mean', width=1.5, line_color="brown", alpha=0.8, source=year, 
         legend_label="IMDb Rating (Mean)")
p.line(x='Year', y='Your Rating_mean', width=1.5, line_color="green", alpha=0.8, source=year, 
         legend_label="My Rating (Mean)")

p.xaxis.axis_label = "Release Year"
p.yaxis.axis_label = "Rating"
p.y_range.start = 5
p.xaxis.major_label_orientation = 3.14/2
p.title.text = "Films (Release Year / Rating)"
p.legend.location = "bottom_left"

output_notebook()
show(p)
```

## En çok hangi oyu kullandım?

```python
df['Your Rating'] = df['Your Rating'].astype(str)

rating_group = df.groupby(('Your Rating'))

grating = ColumnDataSource(rating_group)

rating_group.describe()
```

```python
p = figure(x_range=rating_group, plot_height=500)

p.vbar(x='Your Rating', top='Position_count', width=0.7, line_color="black", color="orange", alpha=0.7, source=grating)

p.xaxis.axis_label = "My Rating"
p.yaxis.axis_label = "Film Count"
p.y_range.start = 0
p.title.text = "Films (My Rating / Film Count)"

output_notebook()
show(p)
```
